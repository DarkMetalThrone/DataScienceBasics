```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create x and y values for plotting
x = np.linspace(-40, 40, 1000)
print(x.size)
y = np.sin(x)
print(y.size)
```

```python
# you have to use matplotlib library to plot any numpy array
plt.plot(x,y)
# we get our familiar sine wave as the graph
```

```python
# A quadratic equation
z = x*x + 5*x + 7
plt.plot(x,z)
```
