# Displot, Kdeplot, Histplot and Rugplots

These are used to do univariate analysis on a continous/numerical data

- Histogram (histplot)
- Kernel Density Estimation (kdeplot)
- Rugplot (displot with rug=True)

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# we plan to use the age column, but this has missing values
# sill the missing values with mean of age
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
# make the changes permanent
```

### Plotting a Displot

```python
# Keeping everything ON
sns.displot(titanic['age'], kde=True, rug=True)
# by default it shows the histogram and the KDE
```

```python
# Plotting only the HISTOGRAM
sns.histplot(titanic['age'], bins=10)
# we can set the bin size manually as well
```

```python
#Plotting only the KDE
sns.kdeplot(titanic['age'])
# gives you the probability of finid the element with that value
```

```python
# Plotting only the Rugplot
sns.displot(titanic['age'], rug=True)
# Rug plot can tell which is the highest occuring number in a bin, if it were a histogram
```

## How do you use a Distplot??

You want to plot the distribution of all the passengers that survived the accident

```python
sns.displot(titanic[titanic['survived']==1]['age'], kde=True)
# A Dataframe of ages of people who survived
sns.displot(titanic[titanic['survived']==0]['age'], kde=True)
# A Dataframe of ages of people who died
```
