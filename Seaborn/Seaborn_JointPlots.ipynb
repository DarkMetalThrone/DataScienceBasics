# Joint Plots

Joint plots creates a multipanel figure that shows both the bivariate(joint) relationship between two variables along with the univariate distribution of each on separate axes

We can use Joint plots in 4 manners:
- ScatterPlots
- Hexbin Plots (Hexagonal Shaped Plot)
- Kdeplots (Kernel Density Estimation)
- RegPlots (Regression Plots)

```python
# import the libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# JointPlot (default = Scatter Plot)
sns.jointplot(x='total_bill', y='tip', data=tips)
# this plots the regular scatterplots and Histograms of the variables passed
```

```python
# Hexbin plot
sns.jointplot(kind='hex', x='total_bill', y='tip', data=tips)
# hexplots are basically 2D histograms
```

```python
# KDE plots
sns.jointplot(kind='kde', x='total_bill', y='tip', data=tips)
# visualise this like the mountain elevation
```

```python
# Regression Plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```
