### Filtering DataFrames

```python
# masking 
# suppose you want the rows with hyderabad as the city
mask = df['city'] == 'Hyderabad'
# we use this boolean series to filter out dataframe
df[mask]
```

```python
# notice how the number of matches is the same as number of rows left
# we can make a function from it
def get_city_match_count(dataframe, city):
    mask = dataframe['city'] == city
    return dataframe[mask].shape[0] # shape[0] will return the number of rows hence the count
```

```python
print(get_city_match_count(df, 'Bangalore'))
# Functions once defined can be used in of the future cells
```

```python
# The filtering criteria could be more complex
# Just create multiple masks
# matched in hyderabad, but only after 2010
mask1 = df['city'] == 'Hyderabad'
mask2 = df['date'] > '2010-01-01'

df[mask1 & mask2].shape[0] # .shpae[0] will return the row count or no of matches
```

### Series in Pandas

- Series' are in the end like key-value pairs
- they have two attributes, index and values
- index doesnt repeat while the values can

```python
myseries = df['winner'].value_counts()
```

```python
print(myseries.index)
print("-------------------------------------------------")
print(myseries.values)
# head() and tail() are still applicable on series
```

```python
# you can access these series similar to lists
myseries['Mumbai Indians']
# will return the value
```

```python
# consider the chance that teams are counted as team1 and team2 in the data
# which team has won the most number of matches??
# you can straight-up add values in two series if you have their indexes
df['team1'].value_counts() + df['team2'].value_counts()
# you can perform all basic mathrmatical operations with these indexes
```

```python
# and all the modular power of OOPs can be leveraged
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values(ascending=False)
```

```python
# you can sort the entire dataframe using this sort()
df.sort_values('city', ascending=False)
# the first parameter is the column which is the criteria
```

```python
# you can make these sorting as permanent - HIGHTLY NOT RECOMMENDED
df2 = df
# use inplace= parameter
```

```python
df2.sort_values(['city','date'], ascending=False, inplace=True)
# now df2 is changed forever
```

```python
# Conditional Sorting
# if for some reason you need one parameter sorted in ascending order and the other in descending order
df.sort_values(['city', 'date'], ascending=[True, False])
# pass both the column names and the sort preference as a list in the method call
```

### Dropping Duplicates

```python
# Duplicates are dropped by the columns
# if you pass 'city as the column then only entries such that all cities are unique will remain
df.drop_duplicates(subset=['city']).shape
# note that the row count has reduced significantly
```

```python
# you can pass drop duplicates on multiple columns
df.drop_duplicates(subset=['city', 'season']).shape
# only unique combinations of 'city' and 'season' are retained
```

```python
# How would you get the season winners of all the years
# you can use drop_duplicates for this purpose
# note that the final match is the last match in the season
# if you drop duplicates and retain the last occurence of 'winner' in every season you have the season champions
df.drop_duplicates('season', keep='last')[['season', 'winner']].sort_values('season')
```

```python
# top5 bats men over the seasons
# import the other dataset
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
# now group by the batsman column in the new dataframe
runs = delivery.groupby('batsman')
# this will create groups based on the names of batsmen from the dataset
```

```python
runs.get_group('V Kohli').shape
# if you do this you'll just get the number of balls played by him
# this is not the right way
```

```python
# you'll have to sum the runs on every delivery
# you can use the .sum() method to get the sum of runs for every batsman
scores = runs['batsman_runs'].sum()
# this will return the sum for every batsman, store it in a new data frame
# you can access the desired batsman with the index
scores['V Kohli']
```

```python
# to get the list of all the batsman
scores.sort_values(ascending=False).head(10)
# this reveals Raina is actually the top scorer for the data we have
```

```python
# find the batsman with the most number of four's
# drop all the entries which are not runs==4
# group by the batman and use count
mask = delivery['batsman_runs'] == 6
new_delivery = delivery[mask]
new_delivery.shape[0] # the number of four's 
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
# count the number of accurences of batsman_runs == 4
# in this case we only have rows where batsman_runs == 4, so we just count them
# dont forget to sort them in descending order
```

```python
# find the top 3 scores by virat kohli against other teams
# group by mask out all the players excpet 'V Kohli'
# opposite team and then with batsman as the index sum the score and sort in descending order
mask = delivery['batsman'] == 'V Kohli'
kohli_delivery = delivery[mask]
kohli_delivery.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# you can create a function based on this logic
def highest_runs_against(batsman_name):
    mask = delivery['batsman'] == batsman_name
    batsman_delivery = delivery[mask]
    # returning the index of the head with one element
    # will return the team against which he has scored the highest runs
    return batsman_delivery.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]

# now putting somebody else's name
print(highest_runs_against('MS Dhoni'))
```
