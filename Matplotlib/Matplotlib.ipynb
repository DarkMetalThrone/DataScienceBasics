# Matplotlib

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Pie Chart

```python
areas = ['Marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
# these can be replaced with any numpy arrays
plt.pie(budget, labels=areas)
# basically call the 'pie function to get a piechart
```

```python
# to show the contents of the pie charts
plt.pie(budget, labels=areas, autopct='%0.2f%%')
# you add the first% to indicate you are formatting the next text
# the %% in the end are the way to show the '%' sign inside the pie chart
```

## Scatter Plot

```python
np.random.seed(2)
x = np.random.randint(10, size=10)
y = np.random.randint(10, size=10)
# create some random arrays
#
plt.scatter(x,y)
plt.xlabel('variable1')
plt.ylabel('variable2')
plt.title('Variable1 Vs Variable2')
```

## 2D plots

```python
# Create some arrays
product_id = np.arange(1, 9)
np.random.seed(3)
online_price = np.random.randint(250, 500, size=8)
offline_price = np.random.randint(250, 500, size=8)
chor_bazaar = np.random.randint(250, 500, size=8)

# plotting a single graph is simple enough
plt.plot(product_id, online_price, color='red', marker='*', markersize=10) 
# color does what it says and marker is pretty self-explainatory
plt.xlabel("ProductID")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

```python
# plotting multiple arrays in the same array
# use the "label=" input to make a legend appear
plt.plot(product_id, online_price, color='red', marker='*', markersize=10, label='Online') 
plt.plot(product_id, offline_price, color='blue', marker='D', markersize=10, label='Offline') 
plt.plot(product_id, chor_bazaar, color='green', marker='+', markersize=10, label='RoadSide')
plt.xlabel("ProductID")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.legend() # now the labels will show up in a legend
plt.show()
```

## Bar Graph

```python
year = np.arange(2012, 2019)
price = np.random.randint(65000, 80000, size=7)
# bar graphs can be plotted with .bar() method
plt.bar(year, price, color='yellow') # for horizontal bar graph use -barh
plt.xlabel('Year')
plt.ylabel("Price(in INR)")
plt.title('YOY of Price')
plt.show()
```
